```python
#Dependencia que utiliza el archivo
import pandas as pd 

df=pd.read_excel("Short_Laprida_Sian_TPINTEGRADOR.xlsx",sheet_name="Hoja1",header=1)
```

```python
#verifir que haya leido correctamente el archivo excel
df.info()
```

# Limpieza de dataframe

```python
#Eliminacion de las columnas que se generaron por error
df=df.drop(columns={"Unnamed: 0","Unnamed: 1"})
```

```python
#verificacion de porque la columna dice que es un objeto cuando deberia de ser un float
df["Promedio diario de desbloqueo de pantalla"].unique()
```

```python
#tranforma los valores de la comuna a numericos para reemplazar los valores que no son numericos por NaN
df["Promedio diario de desbloqueo de pantalla"]=pd.to_numeric( df["Promedio diario de desbloqueo de pantalla"],errors="coerce")
```

```python
#reemplaza los valores NaN por el promedio de la columna
df["Promedio diario de desbloqueo de pantalla"]=df["Promedio diario de desbloqueo de pantalla"].fillna(df["Promedio diario de desbloqueo de pantalla"].mean())
```

```python
#verificacion que se haya hecho los cambios correctos
df["Promedio diario de uso de pantalla"].unique()
```

```python
#Verificacion si hay otra columnas que tenga valores vacios
df.info()
```

# Tabla de frecuencias


```python

bins=[0,2,4,6,8,10,12]
labels=["0-2","2-4","4-6","6-8","8-10","10-12"] 

df['Horas_intervalo'] = pd.cut(df["Horas promedio en redes sociales"], bins=bins, labels=labels, right=False)
```

```python
#verificacion de que se haya hecho correctamente la columna de intervalos
df["Horas_intervalo"].head()
```

```python
#calculo de frecuencias

frecuencia_abs=df["Horas_intervalo"].value_counts().sort_index()
frecuencia_rel=(frecuencia_abs/len(df))*100
frecuencia_acum=frecuencia_abs.cumsum()
frecuencia_acum_rel=frecuencia_rel.cumsum()
```

```python
#Tabla de frecuencias
tabla_frecuencia=pd.DataFrame({"Frecuencia absoluta":frecuencia_abs,
                            "Frecuencia relativa":frecuencia_rel,
                            "Frecuencia acumulada":frecuencia_acum,
                            "Frecuencia acumulada relativa":frecuencia_acum_rel})

print(tabla_frecuencia)
```

# Estadistica descrictiva

```python
media=df["Horas promedio en redes sociales"].mean()
print("media:",media)
mediana=df["Horas promedio en redes sociales"].median()
print("mediana:",mediana)
moda=df["Horas promedio en redes sociales"].mode()[0]
print("moda:",moda)
varianza=df["Horas promedio en redes sociales"].var()
print("varianza:",varianza)
desviacion_estandar=df["Horas promedio en redes sociales"].std()
print("desviación estándar:",desviacion_estandar)
rango=df["Horas promedio en redes sociales"].max()-df["Horas promedio en redes sociales"].min()
print("rango:",rango)
```

```python
#Hallando los valores de Q1 y Q3 para calcular el rango intercuartilico
Q1=df["Horas promedio en redes sociales"].quantile(0.25)
print("Q1:",Q1)
Q3=df["Horas promedio en redes sociales"].quantile(0.75)
print("Q3:",Q3)
IQR=Q3-Q1  
print("IQR:",IQR)
```

```python
#Determinar el sesgo de la distribucion
if media>mediana:
    sesgo="positivo"
elif media<mediana:
    sesgo="negativo"
else:
    sesgo="simétrico"

print(sesgo)
```

```python
#Interpretación de los resultados

print(f"la distribucion presenta un sesgo:{sesgo}")
print("la media y mediana indican que los datos estan distribuidos de manera desigual")
print(f"la desviacion estandar muestra que los datos se alejan de la media en promedio {desviacion_estandar:.4f} unidades")
print(f"los datos se encuentran en un rango de {rango} unidades")
print(f"El {frecuencia_rel.max():.1f}% de los estudiantes se encuentra en el intervalo {tabla_frecuencia["Frecuencia relativa"].idxmax()} horas.")
print(f"El {frecuencia_acum_rel.iloc[-1]:.1f}% de los estudiantes usa entre 0 y 12 horas de redes sociales.")
```
