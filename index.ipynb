```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Short_Laprida_Sian_TPINTEGRADOR.xlsx",sheet_name="Hoja1",header=1)
```

```python
df.info()
```

```python
df=df.drop(columns={"Unnamed: 0","Unnamed: 1"})
```

```python
df["Promedio diario de desbloqueo de pantalla"].unique()
```

```python
df["Promedio diario de desbloqueo de pantalla"]=pd.to_numeric( df["Promedio diario de desbloqueo de pantalla"],errors="coerce")
```

```python
df["Promedio diario de desbloqueo de pantalla"]=df["Promedio diario de desbloqueo de pantalla"].fillna(df["Promedio diario de desbloqueo de pantalla"].mean())
```

```python
df.info()
```

```python
df["Promedio diario de uso de pantalla"].unique()
```

```python
bins=[0,2,4,6,8,10,12]
labels=["0-2","2-4","4-6","6-8","8-10","10-12"] 

df['Horas_intervalo'] = pd.cut(df['Promedio diario de uso de pantalla'], bins=bins, labels=labels, right=False)
```

```python
df["Horas_intervalo"].head()
```

```python
frecuencia_abs=df["Horas_intervalo"].value_counts().sort_index()
```

```python
frecuencia_rel=(frecuencia_abs/len(df))*100
frecuencia_acum=frecuencia_abs.cumsum()
frecuencia_acum_rel=frecuencia_rel.cumsum()
```

```python
tabla_frecuencia=pd.DataFrame({"Frecuencia absoluta":frecuencia_abs,
                            "Frecuencia relativa":frecuencia_rel,
                            "Frecuencia acumulada":frecuencia_acum,
                            "Frecuencia acumulada relativa":frecuencia_acum_rel})

print(tabla_frecuencia)
```
